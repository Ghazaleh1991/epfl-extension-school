```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MaxNLocator
%matplotlib inline
data = pd.read_csv('data.csv')
data.dropna(inplace=True)
data.at[129,'Region'] = 'Eastern Europe'
data
```

```python
top10_happiest = data.head(10)
plot = top10_happiest.plot.barh(y = 'Happiness Score', color='Blue')
plot.set_xlabel('Happiness Score')
plot.set_ylabel('Country')
plot.set_title('Top 10 Happiest Countries in the World')
plot.invert_yaxis()
plot.set_yticklabels(data.head(10)['Country'])
plot.set_xlim(0, 8)
plot.legend_.remove()
```

```python
Economy_score = top10_happiest['Economy']
Family_score = top10_happiest['Family']
Health_score = top10_happiest['Health']
Freedom_score = top10_happiest['Freedom']
Generosity_score = top10_happiest['Generosity']
Corruption_score = top10_happiest['Corruption']
Dystopia_score = top10_happiest['Dystopia']
ind = np.arange(10)
plt.barh(ind, Economy_score, label = 'Economy', color = 'red')
plt.barh(ind, Family_score, left = Economy_score, label = 'Family', color = 'blue')
plt.barh(ind, Health_score, left = Economy_score + Family_score, label = 'Health', color = 'green')
plt.barh(ind, Freedom_score, left = Economy_score + Family_score + Health_score, label = 'Freedom', color = 'black')
plt.barh(ind, Generosity_score, left = Economy_score + Family_score + Health_score + Freedom_score, label = 'Generosity', color = 'purple')
plt.barh(ind, Corruption_score, left = Economy_score + Family_score + Health_score + Freedom_score + Generosity_score, label = 'Corruption', color = 'yellow')
plt.barh(ind, Dystopia_score, left = Economy_score + Family_score + Health_score + Freedom_score + Generosity_score + Corruption_score, label = 'Dystopia', color = 'brown')
plt.xlim(0, 8)
plt.yticks(ind, data.head(10)['Country'])
plt.title('Contribution of Happiness Parameters for Top 10 Happiest Countries in the World')
plt.xlabel('Score')
plt.ylabel('Country')
plt.gca().invert_yaxis()
plt.legend()
```

```python
af_Economy_score = data[data['Region'] == 'Africa']['Economy'].reset_index(drop = True)
af_Family_score = data[data['Region'] == 'Africa']['Family'].reset_index(drop = True)
af_Health_score = data[data['Region'] == 'Africa']['Health'].reset_index(drop = True)
af_Freedom_score = data[data['Region'] == 'Africa']['Freedom'].reset_index(drop = True)
af_Generosity_score = data[data['Region'] == 'Africa']['Generosity'].reset_index(drop = True)
af_Corruption_score = data[data['Region'] == 'Africa']['Corruption'].reset_index(drop = True)
af_Dystopia_score = data[data['Region'] == 'Africa']['Dystopia'].reset_index(drop = True)
af_ind = np.arange(len(af_Economy_score.index))
plt.barh(af_ind, af_Economy_score, label = 'Economy', color = 'red')
plt.barh(af_ind, af_Family_score, left = af_Economy_score, label = 'Family', color = 'blue')
plt.barh(af_ind, af_Health_score, left = af_Economy_score + af_Family_score, label = 'Health', color = 'green')
plt.barh(af_ind, af_Freedom_score, left = af_Economy_score + af_Family_score + af_Health_score, label = 'Freedom', color = 'black')
plt.barh(af_ind, af_Generosity_score, left = af_Economy_score + af_Family_score + af_Health_score + af_Freedom_score, label = 'Generosity', color = 'purple')
plt.barh(af_ind, af_Corruption_score, left = af_Economy_score + af_Family_score + af_Health_score + af_Freedom_score + af_Generosity_score, label = 'Corruption', color = 'yellow')
plt.barh(af_ind, af_Dystopia_score, left = af_Economy_score + af_Family_score + af_Health_score + af_Freedom_score + af_Generosity_score + af_Corruption_score, label = 'Dystopia', color = 'brown')
plt.xlim(0, 8)
plt.gca().invert_yaxis()
plt.yticks(af_ind, data[data['Region'] == 'Africa']['Country'])
plt.title('Contribution of Happiness Parameters for African Countries')
plt.xlabel('Score')
plt.ylabel('Country')
```
