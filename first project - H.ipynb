```python
import pandas as pd
import numpy as np
import random

data = pd.read_csv('data.csv')
data.dropna(inplace=True)
data.at[129,'Region'] = 'Eastern Europe'
data
```

```python
data.reset_index(drop = True, inplace = True)
region_list = ['Africa', 'Asia-Pacific', 'Eastern Europe', 'Latin America', 'North America', 'Western Europe']
cr_mat = np.zeros((len(data.index), len(region_list)))
for i in range(len(data.index)):
    for j in range(len(region_list)):
        if data.loc[i]['Region'] == region_list[j]:
            cr_mat[i][j] = 1
            break

cr_mat
```
