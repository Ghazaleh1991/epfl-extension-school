```python
###Part A - Continue##
#If the values in 'additives' column include the string "exists", it identifies an additive
#for example in [ citric-acid -> en:e330 -> exists -- ok ] we should delete the first part as '[ ' and the second part which starts from the first '->'
#first, let's split the ingredients in each row of the 'additives' column
for index_no in food_facts.index:
    additive_list = []
    additive_counter = 0
    if food_facts['additives_n'].at[index_no] != 'Not Mentioned' and int(food_facts['additives_n'].at[index_no]) > 0:
        splitted = []
        splitted = food_facts['additives'].at[index_no].split('  ]  [ ')
        #Now let's omit the opening bracket from the first and the closing bracket from the last data in "splitted"
        splitted[0] = splitted[0].replace(' [ ', '')
        splitted[len(splitted) - 1] = splitted[len(splitted) - 1].replace('  ] ', '')
        #Now that the rows in 'additives' column are splitted, let's check if the listed ingredient is an additive
        for additive_index in np.arange(len(splitted)):
            if 'exists' in splitted[additive_index]:
                additive_counter = additive_counter + 1
                #let's extract the additive
                for character_index in np.arange(len(splitted[additive_index]) - 2):
                    if splitted[additive_index][character_index:character_index + 2] == '->':
                        additive_list.append(splitted[additive_index][0:character_index - 1])
                        break
            if additive_index == len(splitted) - 1:
                #make the new data for the row 
                new_data = ''
                for selected_additive_index in np.arange(len(additive_list)):
                    new_data = new_data + additive_list[selected_additive_index]
                    if selected_additive_index == len(additive_list) - 1:
                        food_facts['additives'].set_value(index_no, new_data)
                    else:
                        new_data = new_data + ', '                                                    
                #check if the number of found additives is equal to the value presented in 'additives_n' column
                if additive_counter != int(food_facts['additives_n'].at[index_no]):
                    #if not, set the number of additives to their actual values
                    food_facts['additives_n'].set_value(index_no, additive_counter)
```

```python
###Part B - 1###
#The Atwater system uses the average values of 4 Kcal/g (16.736 kJ/g) for protein
#4 Kcal/g (16.736 kJ/g) for carbohydrate, and 9 Kcal/g (37.656 kJ/g) for fat for calculation of energy content of a food.
#Based on this informaton, the difference between the presented and calculated energy of each food is presented in a new column
#Finally, the mean and highest level of energy difference is calculated and presented
n_food_facts = food_facts[food_facts['energy_100g'] != 'Not Mentioned']
n_food_facts = n_food_facts[n_food_facts['proteins_100g'] != 'Not Mentioned']
n_food_facts = n_food_facts[n_food_facts['carbohydrates_100g'] != 'Not Mentioned']
n_food_facts = n_food_facts[n_food_facts['fat_100g'] != 'Not Mentioned']
n_food_facts['energy_difference'] = (n_food_facts['energy_100g'].astype(float) - n_food_facts['proteins_100g'].astype(float) * 16.736 - n_food_facts['carbohydrates_100g'].astype(float) * 16.736 - n_food_facts['fat_100g'].astype(float) * 37.656).astype(int)
n_food_facts[['energy_100g', 'ingredients_text', 'proteins_100g', 'carbohydrates_100g', 'fat_100g', 'energy_difference']]
```
